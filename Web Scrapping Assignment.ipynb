Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of 
release).

```python
page= requests.get("https://www.imdb.com/chart/top")
page
```

```python
Soup= BeautifulSoup(page.content)
Soup
```

```python
Year = []

for i in year:
    Year.append(i.text)

Year
```

```python
ratings= Soup.find_all("td", class_="ratingColumn imdbRating")
ratings
```

```python
Ratings = []

for i in ratings:
    Ratings.append(i.text.replace ("\n"," "))

Ratings
```

```python
import pandas as pd 
```

```python
movies= pd.DataFrame({})
movies['Movie_title']= Movie_title[:100]
movies['Ratings']= Ratings[:100]
movies['Year']= Year[:100]

movies
```

Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
page= requests.get("https://bookpage.com/")
page
```

```python
soup=BeautifulSoup(page.content, "html.parser")
soup
```

```python
url_tag= soup.find_all("h4",class_="italic")
url_tag
```

```python
urls= []
for i in url_tag:
    for j in i.find_all("a", href=True):
        urls.append(j["href"])
        
urls
```

```python
review = []
for i in urls[:5]:
    book= requests.get("https://bookpage.com" + i)
    soup=BeautifulSoup(book.content, "html.parser")
    
    rev= soup.find("div", class_="article-body")
    review.append(rev.text)
    
review
```

```python
len(review)
```

```python
Book_name = []
for i in urls[:5]:
    name= requests.get("https://bookpage.com" + i)
    soup=BeautifulSoup(name.content, "html.parser")
    
    names= soup.find("h4", class_="italic")
    Book_name.append(names.text.replace("\n"," "))
    
Book_name
```

```python
Author_name = []
for i in urls[:5]:
    author= requests.get("https://bookpage.com" + i)
    soup=BeautifulSoup(author.content, "html.parser")
    
    author_names= soup.find("p", class_="sans bold")
    Author_name.append(author_names.text.replace("\n"," "))
    
Author_name
```

```python
Genre = []
for i in urls[:5]:
    book= requests.get("https://bookpage.com" + i)
    soup=BeautifulSoup(book.content, "html.parser")
    
    genre_names= soup.find("p", class_="genre-links hidden-phone")
    Genre.append(genre_names.text)
    
Genre
```

```python
period = tonight.find("p",class_ = "period-name").get_text()
short_Desc = tonight.find("p",class_ = "short-desc").get_text()
temp = tonight.find("p",class_ = "temp temp-low").get_text()

print(period)
print(short_Desc)
print(temp)
```

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
```

Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
company name, CTC, and apply date.


```python
page= requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
job_titles = []
company_names= []
job_info= []
CTC = []
apply_date = []
```

```python
titles= soup.find_all("div",class_="heading_4_5 profile")

for i in titles:
    job_titles.append(i.text.strip())
  
    
job_titles
```

```python
company= soup.find_all("a",class_="link_display_like_text")

for i in company:
    company_names.append(i.text.strip())
  
    
company_names
```

```python
info= soup.find_all("div",class_="item_body")
info
```

```python
job_info= []

for i in info:
    job_info.append(i.text.replace("\n","").replace("\xa0",""))
    
job_info
```

```python
CTC = []
for i in range(1,len(job_info),3):
    CTC.append(job_info[i].strip())
    
CTC
```

```python
apply_date = []
for i in range(2,len(job_info),3):
    apply_date.append(job_info[i].strip())
    
apply_date
```

```python
Job_Postings= pd.DataFrame({})
Job_Postings["job_titles"]= job_titles
Job_Postings["company_names"]= company_names
Job_Postings["CTC"]= CTC
Job_Postings["apply_date"]=apply_date

Job_Postings
```
