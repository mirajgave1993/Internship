1. Scrape the details of most viewed videos on YouTube from Wikipedia:

You need to find following details:
    
A) Rank

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.maximize_window()
time.sleep(2)
```

```python
#creating the empty list
Rank=[]
Name=[]
Artist=[]
Upload=[]
Views=[]
```

```python
#scraping the Rank
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]"):
    Rank.append(i.text)

#scraping the Name
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]"):
    Name.append(i.text.split('[')[0])

#scraping the Artist
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]"):
    Artist.append(i.text)

#scraping the upload date
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]"):
    Upload.append(i.text)

#scraping the number of views
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]"):
    Views.append(i.text)
```

```python
driver.get('https://github.com/')
driver.maximize_window()
```

```python
#clicking on trending
trending=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
driver.get(trending.get_attribute('href'))
```

```python
#creating the empty list
name=[]
url=[]
desc=[]
count=[]
lang=[]
```

```python
#getting the url of all the repository
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(i.get_attribute('href'))

#scraping the name
for i in driver.find_elements_by_xpath("//span[@class='text-normal']"):
    name.append(i.text.replace(' /',''))
                
for i in driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']"):
    desc.append(i.text)
 
 #looping through urls to scrape other details

for i in url:
    driver.get(i)

#scraping for count
    count1=driver.find_elements_by_xpath("//div[@class='BorderGrid-cell']/h2/a/span")
    try:
        count.append((count1[-1].text))
    except:
        count.append('-')
        
#scraping for language
    language= driver.find_elements_by_xpath("//ul[@class='list-style-none']")
    try:
        lang.append((language[-1].text.replace('\n',' ')))
    except:
        lang.append('-')
```

```python
#creating the dataframe
Github={'Repository Title':name[0:20],'Repository description':desc[0:20],'Contributors count':count[0:20],'Language used':lang[0:20]}
Github=pd.DataFrame(data=Github)
Github
```
